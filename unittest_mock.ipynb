```python
from unittest.mock import patch
import unittest
```

```python
class People:
    def __init__(self):
        pass
    
    def falar_horas(self):
        from datetime import datetime
        return datetime.now()
        
    def _gerar_numero(self):
        from random import randint
        return randint(0, 10)
    
    def resolver_multiplicacao(self):
        return self._gerar_numero() * 2
```

```python
# @patch.object(People, '_gerar_numero', return_value=11)
@patch('random.randint', return_value=121)
class TestAllPeople(unittest.TestCase):
    
    def setUp(self) -> None:
        self.people = People()
    
    def test_falar_horas(self, mock_output):
        print(self.people.falar_horas())
        self.assertTrue(True)
        
    def test_gerar_numero(self, mock_output):
        print(self.people._gerar_numero())
        self.assertTrue(True)

    def test_resolver_multiplicacao(self, mock_output):
        output = self.people.resolver_multiplicacao()
        print(output)
        self.assertEqual(output, 242)
```

```python
suite = unittest.TestLoader().loadTestsFromTestCase(TestAllPeople)
unittest.TextTestRunner(verbosity=2).run(suite)
```
